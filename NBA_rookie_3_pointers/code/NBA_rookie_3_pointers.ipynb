```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data

```python
data = {"player": ["Keegan Murray",
                   "Donovan Mitchell",
                   "Damian Lillard",
                   "Saddiq Bey",
                   "Anthony Edwards",
                   "Luka Doncic",
                   "Landry Shamet",
                   "Stephen Curry",
                   "Kyle Kuzma",
                   "Rudy Fernandez"],
        "number": [188,187,185,175,171,168,167,166,159,159],
        "team": ["Kings, 2022-23",
                 "Jazz, 2017-18",
                 "Trail Blazers, 2012-13",
                 "Pistons, 2020-21",
                 "Timberwolves, 2020-21",
                 "Mavericks, 2018-19",
                 "76ers/Clippers, 2018-19",
                 "Warriors, 2009-10",
                 "Lakers, 2017-18",
                 "Trail Blazers, 2008-09"]}

data = pd.DataFrame.from_dict(data)
data
```

```python
x = [a+ "\n("+b+")" for a,b in zip(data["player"],data["team"])]
y = data["number"]
colors = ["Purple","Black","Red","Blue","Grey","DarkBlue","Silver","Yellow","Gold","Red"]

plt.figure(figsize=(12,6))
bar = plt.bar(x,y,color=colors)

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, f'{height:.0f}', ha='center', va='bottom',fontsize=16)

plt.ylim([0,210])

plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=14)

plt.title('Number of 3-pointers made in rookie NBA season as of 3/29/2023', fontsize=18)
plt.xlabel('Player', fontsize=16)
plt.ylabel('Number of 3-pointers', fontsize=16)

plt.savefig("../images/NBA_rookie_3_pointers.png", bbox_inches="tight")
# plt.show()
```
